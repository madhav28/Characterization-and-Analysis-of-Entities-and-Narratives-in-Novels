```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("coherence_scores.csv")

max_index = df['Coherence Score'].idxmax()
max_value = df['Coherence Score'].max()

plt.figure(figsize=(10, 6))
plt.plot(df['Iteration'], df['Coherence Score'], marker='o', label='Coherence Score')

plt.scatter(df['Iteration'][max_index], max_value, color='red', s=100, zorder=5, label=f'Highest: {max_value}', marker='*')

plt.xlabel('Iteration')
plt.ylabel('Coherence Score')
plt.title('Coherence Score vs Iteration')
plt.legend()
plt.savefig("coherence_scores")
plt.show()
```

```python
df = pd.read_csv("doc_info.csv")

freq = df['Narrative'].value_counts().head(10)

plt.figure(figsize=(10, 6))
freq.plot(kind='barh', color='skyblue', edgecolor='black')

plt.xlabel('Frequency')
plt.ylabel('Narrative')
plt.title('Top 10 Frequent Narratives (Entire Corpus)')
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.savefig("top_10_entire")
plt.show()
```

```python
df = pd.read_csv("secret_adversary_doc_info_new.csv")

df['Narrative Theme'] = df['Narrative Theme'].replace(" Sabin's search for the Wolfenden family and their connection to the Countess of Whitlett",
                                                      " Sabin's search for the Wolfendens and the Countess of Whitlett")
df['Narrative Theme'] = df['Narrative Theme'].replace(' Investigation and mystery solving in Agatha Christie\'s "The Secret Adversary"',
                                                      'Investigation and mystery solving')
freq = df['Narrative Theme'].value_counts().head(10)

plt.figure(figsize=(10, 6))
freq.plot(kind='barh', color='skyblue', edgecolor='black')

plt.xlabel('Frequency')
plt.ylabel('Narrative')
plt.title('Top 10 Frequent Narratives (Secret Adversary)')
plt.gca().invert_yaxis()  
plt.tight_layout()
plt.savefig("top_10_secret_adversary")
plt.show()
```
