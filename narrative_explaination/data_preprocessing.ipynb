```python
import pandas as pd
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
doc_data = pd.read_csv("data/doc_info.csv")
doc_data
```

```python
doc_data[['Author', 'Book']] = doc_data['Book Name'].str.extract(r"^(.*?)___(.*)\.txt$")
doc_data.drop('Book Name', axis=1, inplace=True)
doc_data
```

```python
authors_to_filter = ['Sir Arthur Conan Doyle']
filtered_data = doc_data[doc_data['Author'].isin(authors_to_filter)]
filtered_data
```

```python
len(filtered_data['Book'].unique())
```

```python
filtered_data = filtered_data.groupby("Book").head(2)
filtered_data
```

```python
filtered_data = filtered_data.drop(columns=["Top_n_words","Probability","Representative_document","Topic","Name","Representation","Representative_Docs"])
filtered_data
```

```python
filtered_data.to_csv(".csv",index=False)
```
